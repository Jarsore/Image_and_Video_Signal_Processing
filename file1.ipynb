```python
#
img1 = cv2.imread("./img/cc1.jpg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

rows,cols,channel = img1.shape

npimg = np.ones(shape = img1.shape,dtype = np.uint8)*255

npimg[:rows//2,:cols//2,:] = 0

ret_and = cv2.bitwise_and(img1,npimg)
ret_or = cv2.bitwise_or(img1,npimg)
ret_not = cv2.bitwise_not(img1)
ret_xor = cv2.bitwise_xor(img1,npimg)

plt.subplot(221)
plt.imshow(ret_and) 
plt.subplot(222) 
plt.imshow(ret_or) 
plt.subplot(223) 
plt.imshow(ret_not) 
plt.subplot(224)
plt.imshow(ret_xor)
```

```python
#
img = cv2.imread('./img/cc1.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

height,width = img.shape[:2]
tx,ty = width/4,height/4
translation_matrix = np.array([
    [1,0,tx],
    [0,1,ty]
],dtype = np.float32)

translated_image = cv2.warpAffine(src = img,M = translation_matrix,dsize = (width,height))

plt.imshow(translated_image)
```

```python
#
import numpy as np
from math import sin
from math import cos
from math import pi

img = cv2.imread('./img/hlw.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

height,width = img.shape[:2]
center = (width/2,height/2)
rotate_matrix = np.array([
    [cos(pi/4),-sin(pi/4),0],
    [sin(pi/4),cos(pi/4),0]
],dtype=np.float32)

rotated_image = cv2.warpAffine(src=img,M=rotate_matrix, dsize=(width, height))
plt.imshow(rotated_image)
```

```python
#
img = cv2.imread('./img/hlw.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
height,width = img.shape[:2]
center = (width/2,height/2)
scale_matrix = np.array([
    [0.5,0,0],
    [0,0.5,0]
],dtype = np.float32)

scale_image = cv2.warpAffine(src = img,M = scale_matrix,dsize = (width,height))
plt.imshow(scale_image)
```

```python
#
img = cv2.imread('./img/hlw.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
height,width = img.shape[:2]
center = (width/2,height/2)
vertical_matrix = np.array([
    [1,0.5,0],
    [0,1,0]
],dtype = np.float32)

horizontal_matrix = np.array([
    [1,0,0],
    [0.5,1,0]
],dtype = np.float32)

vertical_image = cv2.warpAffine(src = img,M = vertical_matrix,dsize = (width,height))
horizontal_image = cv2.warpAffine(src = img,M = horizontal_matrix,dsize = (width,height))
plt.subplot(121)
plt.imshow(vertical_image)
plt.subplot(122)
plt.imshow(horizontal_image)
```

```python
#
img = cv2.imread('./img/swk.jpg') 
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
height,width = img.shape[:2]

nearest_img=cv2.resize(img,(2*width,2*height),cv2.INTER_NEAREST) 
linear_img=cv2.resize(img,(2*width,2*height),cv2.INTER_LINEAR)
cubic_img=cv2.resize(img,(2*width,2*height),cv2.INTER_CUBIC)
area_img=cv2.resize(img,(2*width,2*height),cv2.INTER_AREA)
lanczos_img=cv2.resize(img,(2*width,2*height),cv2.INTER_LANCZOS4)
                       
plt.subplot(231) 
plt.imshow(img)
plt.title("src image") 
plt.subplot(232)
plt.imshow(nearest_img)
plt.title("Nearest interpolation") 
plt.subplot(233)
plt.imshow(linear_img)
plt.title("Linear interpolation") 
plt.subplot(234)
plt.imshow(cubic_img)
plt.title("cubic interpolation") 
plt.subplot(235)
plt.imshow(area_img)
plt.title("Area interpolation") 
plt.subplot(236)
plt.imshow(lanczos_img)
plt.title("Lanczos interpolation")
```
